```python
import matplotlib.pyplot as plt
import cv2 as cv

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
font1 = {
      'color':  'darkblue',
      'weight': 'bold',
      'size': 16}
```

```python
jaum = {
	'ㅁ' : (90, 70),
	'ㄴ' : (140, 75),
	'ㅇ' : (250, 95),
	'ㅂㅃㅍ' : (55, 215),
	'ㄷㄸㅌㄹ' : (115, 160),
	'ㅅㅆ' : (160, 140),
	'ㅈㅉㅊ' : (220, 140),
	'ㄱㄲㅋ' : (300, 160),
	'ㅎ' : (335, 300)
}

moum = {
	'ㅣ' : (115, 160),
	'ㅔ' : (125, 190),
	'ㅐ' : (135, 220),
	'ㅟ' : (165, 170),
	'ㅚ' : (175, 200),
	'ㅏ' : (215, 235),
	'ㅡ' : (225, 160),
	'ㅜ' : (265, 170),
	'ㅗ' : (260, 195),
	'ㅓ' : (255, 220),
}
```

```python
jaumPoses = {}
for key, value in jaum.items():
	if len(key) == 1:
		jaumPoses[key] = value
	else:
		for k in key:
			jaumPoses[k] = value

moumPoses = moum.copy()

print(jaumPoses, moumPoses)
```

```python
originalImg = cv.imread('originalimg.png')

fig, ax = plt.subplots()
ax.axis('off')
for key, value in jaumPoses.items():
	tmp = plt.Circle(value, 5, color='b')
	ax.text(value[0], value[1], key)#@, fontdict=font1)
	ax.add_patch(tmp)

for key, value in moumPoses.items():
	tmp = plt.Circle(value, 5, color='b')
	ax.text(value[0], value[1], key)
	ax.add_patch(tmp)
ax.imshow(originalImg[:,:,::-1])
```
